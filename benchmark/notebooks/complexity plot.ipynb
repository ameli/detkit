```python
import numpy
import matplotlib.pyplot as plt
```

```python
def plot_flops(ax, scale=1):

    r = numpy.linspace(0, 1)
    a = 1  # solving triangular system
    b = 2  # matrix multiplication.
    g1 = 0.5  # set 0.5 for grammian matrix multiplication, set 1 for generic matrix multiplication
    g2 = 1  # set 0.5 for grammian matrix multiplication, set 1 for generic matrix multiplication
    c = 2.0/3.0  # cholesky
    plogdet = 1

    y11 = c + a*r + g1*b*r**2 + c*r**3 + plogdet*(b*r**2 + c*r**3)
    y12 = c + a*r + g2*b*r**2 + c*r**3 + plogdet*(b*r**2 + c*r**3)
    y2 = 2*c + 2*a*r + 2.0000*b*r**2 + 2*c*r**3 + plogdet*(b*r**2 + c*r**3)
    y3 = 2*c + 2*b*r
    y41 = 2*c + 2.0*b*r + a*r**2 + g1*b*r**2 + c*r**3
    y42 = 2*c + 2.0*b*r + a*r**2 + g2*b*r**2 + c*r**3

    ax.plot(r, y11, color='olive')
    ax.plot(r, y12, '--', color='olive')
    ax.plot(r, y2, color='yellowgreen')
    ax.plot(r, y3, color='maroon')
    ax.plot(r, y41, color='indianred')
    ax.plot(r, y42, '--', color='indianred')

fig, ax = plt.subplots(figsize=(5.2, 4))
plot_flops(ax, scale=1)
ax.set_xlim([0, 1])
ax.set_ylabel('Process Time')
#plt.xscale('log')
#plt.yscale('log')
```
